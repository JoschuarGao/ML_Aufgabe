```python
class ResNetDataset(Dataset):
    def __init__(self, data_path, phase, train_size=256, crop_size=224) -> None:
        super().__init__()
        self.phase = phase

        object_names = ['Bicycle', 'Bridge', 'Bus', 'Car', 'Chimney', 'Crosswalk', 'Hydrant', 'Motorcycle', 'Other', 'Palm', 'Stair', 'Traffic_Light']

        self.files = glob.glob(data_path+object_names[0]+'/*.png', recursive=True)
        # self.files = glob.glob('/content/train_val_data/' + object_names[0] + '/*.png', recursive=True)
        self.labels = np.zeros(len(self.files), np.int32)

        for i, name in enumerate(object_names[1:]):
            self.files = self.files+glob.glob(data_path+name+'/*.png', recursive=True)
            self.labels = np.concatenate((self.labels, np.zeros(len(self.files)-len(self.labels), np.int32)+i+1), 0)


        self.files = np.array(self.files)
        if phase == 'train':
            index = []
            for i in range(len(self.files)):
                if i%10 == 0:
                    continue
                index.append(i)

            self.files = self.files[index]
            self.labels = self.labels[index]

        if phase == 'val':
            self.files = self.files[::10]
            self.labels = self.labels[::10]

        self.tfs_train = tfs.Compose([tfs.Resize((train_size,train_size), antialias=True),
                                      tfs.RandomCrop((crop_size,crop_size)),
                                      tfs.RandomHorizontalFlip(0.5),
                                      tfs.ColorJitter(brightness=[0.6, 1.4], contrast=[0.6, 1.4], saturation=[0.6, 1.4], hue=[0, 0]),
                                      tfs.RandomRotation(30),
                                      #tfs.Normalize(mean=(0.485, 0.456, 0.406), std=(0.229, 0.224, 0.225)),
                                      tfs.ToTensor()])
        self.tfs_val = tfs.Compose([tfs.Resize((crop_size,crop_size)),  tfs.ToTensor()])

    def __len__(self):
        return len(self.files)

    def __getitem__(self, idx):
        image = Image.open(self.files[idx]).convert('RGB')
        if self.phase=='train':
            image = self.tfs_train(image)
        else:
            image = self.tfs_val(image)
            #image = torch.stack([image,  F.hflip(image)])

        #image = cv2.imread(self.files[idx])
        #if self.phase == 'train':
        #    image = cv2.resize(image, [160,160])
        #    i = random.randint(0,32)
        #    j = random.randint(0,32)
        #    image = image[i:i+128, j:j+128]
        #else:
        #image = cv2.resize(image, [128,128])
        #image = image.transpose((2, 0, 1))[::-1]  # HWC to CHW, BGR to RGB
        #image = image.astype(np.float32)/255.0

        label = np.zeros(12, np.float32)
        label[self.labels[idx]] = 1
        return image, label

```

```python
epochs = 100
batch_size = 72
num_workers = 0
learning_rate = 0.01
learning_rate_final = 0.001
data_path = 'train_val_data/'
```

```python
train_data = ResNetDataset(data_path=data_path, phase='train')
train_dataloader = DataLoader(train_data, batch_size=batch_size, shuffle=True, num_workers=num_workers, pin_memory=True)

val_data = ResNetDataset(data_path=data_path, phase='val')
val_dataloader = DataLoader(val_data, batch_size=batch_size, shuffle=False, num_workers=num_workers, pin_memory=True)
```

```python
device = torch.device('cuda:0')
```

```python
model = models.resnet50(weights = 'ResNet50_Weights.IMAGENET1K_V2')
model.fc = torch.nn.Linear(in_features=2048, out_features=12)
model = model.to(device)

optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
```

```python
# 学习率衰减策略
lf = lambda x: (1 - x / epochs) * (1.0 - learning_rate_final) + learning_rate_final
scheduler = torch.optim.lr_scheduler.LambdaLR(optimizer, lr_lambda=lf)

loss_function = torch.nn.CrossEntropyLoss()
softmax = torch.nn.Softmax(dim = 1)
```

```python
for epoch_i in range(epochs):
    model.train()
    loss_sum = 0
    for i, (image, label) in enumerate(train_dataloader):
        image, label = image.to(device), label.to(device)
        optimizer.zero_grad()
        output = model(image)
        loss = loss_function(output, label)
        loss_sum = loss_sum+loss.item()
        if (i+1)%100 == 0:
            lr = [x['lr'] for x in optimizer.param_groups]
            print('Train Epoch:', epoch_i, 'Step:', i, 'Loss:', loss_sum/100, 'Learning Rate:', lr)
            loss_sum = 0
        loss.backward()
        optimizer.step()
    scheduler.step()

    model.eval()
    loss_sum = 0
    correct_num = 0
    sample_num = 0
    best_loss = float('inf')
    best_accuracy = 0
    for i, (image, label) in enumerate(val_dataloader):
        image, label = image.to(device), label.to(device)
        #image = image.flatten(0,1)

        sample_num = sample_num + image.shape[0]
        output = model(image)
        #output = output[1::2]+output[1::2]

        loss = loss_function(output, label)
        loss_sum = loss_sum+loss.item()*image.shape[0]

        output = output.detach().cpu().numpy()
        label = label.cpu().numpy()

        pred = np.argmax(output, axis=1)
        gt = np.argmax(label, axis=1)
        correct_num = correct_num+np.sum(pred == gt)
    if loss_sum < best_loss:
        best_loss = loss_sum
        torch.save(model, 'best_loss.pt')
    if correct_num>best_accuracy:
        best_accuracy = correct_num
        torch.save(model, 'best_accuracy.pt')
    print('Val Epoch:', epoch_i, 'Step:', i, 'Loss:', loss_sum/sample_num, 'Accuracy:', correct_num/sample_num)
```
